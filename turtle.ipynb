```python
from heeps.util.coord import polar_coord
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
```

```python
# create lambda functions for fourier transforms
fft = lambda x: np.fft.fftshift(np.fft.fft2(np.fft.fftshift(x)))
ifft = lambda x: np.fft.fftshift(np.fft.ifft2(np.fft.fftshift(x)))
```

```python
nimg = 2**10
nlines = 10
r, t = polar_coord(nimg)
agpm = np.floor(r*nlines) % 2
plt.imshow(agpm, origin=True)
fits.writeto('agpm.fits', agpm, overwrite=True)
```

```python
fagpm = abs(fft(agpm))
plt.imshow(np.log10(fagpm), origin=True)
fits.writeto('fagpm.fits', fagpm, overwrite=True);
```

```python
l = 6
m = l-2
a = 2*np.pi/l
ud = np.floor(t/a) % 2 > 0

t4 = abs(t*m/np.pi)
t4[t4>1] = (2 - t4[t4>1]) 
sgv4 = np.floor((r + t4**2)*nlines) % 2
plt.imshow(sgv4, origin=True)
fits.writeto('sgv4.fits', sgv4, overwrite=True)
```

```python
fsgv4 = abs(fft(sgv4))
plt.imshow(np.log10(fsgv4), origin=True)
fits.writeto('fsgv4.fits', fsgv4, overwrite=True);
```
